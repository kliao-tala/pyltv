```python
from models import AutoRegression, PowerSlope

import pandas as pd
from dbm import DBM
```

### Load Data

```python
db = DBM(user='kenny.liao')
```

```python
# pull most recent
# ke_data = db.get_market_data(market='ke', days_before=0)
# ph_data = db.get_market_data(market='ph', days_before=0)
mx_data = db.get_market_data(market='mx', days_before=0)
```

### Generate Model

```python
# generate models
# ke = AutoRegression(ke_data, market='ke')
# ph = AutoRegression(ph_data, market='ph')
mx = AutoRegression(mx_data, market='mx')
```

### Generate Forecasts

```python
mx.forecast = mx.forecast_data(mx.data, min_months=1, n_months=50)
```

### Prepare Data for Upload

```python
df = mx.forecast.copy()

# change first loan disbursement month to datetime
df['First Loan Local Disbursement Month'] = pd.to_datetime(df['First Loan Local Disbursement Month']).dt.date
```

### Load data

```python
mx.forecast.to_sql(name='ltv_mx', con=db.engine.connect(), schema='FINANCE_MX', if_exists='replace', index=False)
```
