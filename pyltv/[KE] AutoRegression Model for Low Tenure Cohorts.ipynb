```python
from models import AutoRegression, PowerSlope
from dbm import DBM

import pandas as pd
import plotly
from plotly import graph_objects as go
```

```python
dbm = DBM(user='kenny.liao')
```

```python
data = dbm.get_market_data(market='ke', days_before=60)
```

```python
market='ke'
ltv_expected='ke_ltv_expected_6-24-22.csv'
backtest_months=3
```

### PowerSlope

```python
m1 = PowerSlope(data, market=market, ltv_expected=ltv_expected)
m1.forecast = m1.forecast_data(m1.data, min_months=2, n_months=50)
m1.backtest, m1.backtest_report = m1.backtest_data(m1.data, hold_months=backtest_months, min_months=2)
```

### AutoRegression

```python
m2 = AutoRegression(data, market=market, ltv_expected=ltv_expected)
m2.forecast = m2.forecast_data(m2.data, min_months=1, n_months=50)
m2.backtest, m2.backtest_report = m2.backtest_data(m2.data, hold_months=backtest_months, min_months=1)
```

### Plots

```python
m1.plot_cohorts('borrower_retention', 'forecast')
```

```python
m2.plot_cohorts('borrower_retention', 'forecast')
```
